```python
import pandas as pd
import numpy as np
```

```python
retail_raw = pd.read_excel('online_retail.xlsx')
```

```python
retail = retail_raw.copy()
```

```python
retail = retail.dropna()
```

```python
retail = retail[retail.Quantity >= 0] 
```

```python
retail.info()
```

```python
retail.CustomerID = retail.CustomerID.apply(lambda x: str(int(x)))
```

```python
retail.head(1)
```

```python
retail['Revenue'] = retail.Quantity * retail.UnitPrice
```

```python
retail.head(3)
```

```python
retail.groupby('CustomerID').aggregate({'Revenue': np.mean, 'CustomerID': np.size})
```

```python
retail.groupby('CustomerID').aggregate({'Revenue': np.mean, 'CustomerID': np.size}).rename(columns={'Revenue':"RevenueAverage", 'CustomerID': 'CustomerIDcount'})
```

```python
retail['InvoiceDateDay'] = retail.InvoiceDate.dt.day
retail.head(2)
```

```python
result = retail.groupby('InvoiceDateDay').aggregate({'Revenue': np.sum})
```

```python
import seaborn as sns
sns.set()
result.plot(kind='bar')
```
