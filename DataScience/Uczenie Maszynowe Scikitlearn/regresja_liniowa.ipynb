```python
import sklearn
```

```python
import numpy as np
X = np.arange(0,50,0.5)
noise = 10 * np.random.randn(100)
y = 2 * X + 100 + noise
X = X.reshape(-1,1)

print('Rozmiar X: ', X.shape)
print('Rozmiar y: ', y.shape)
```

```python
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.25)

print('X_train: ', X_train.shape)
print('X_test: ', X_test.shape)
print('y_train: ', y_train.shape)
print('y_test: ', y_test.shape)
```

```python
import matplotlib.pyplot as plt
import seaborn as sns
sns.set()

plt.figure(figsize=(10,6))
plt.scatter(X_train, y_train, label='Dane treningowe')
plt.scatter(X_test, y_test, label='Dane testowe')
plt.xlabel('X')
plt.ylabel('y')
plt.legend()
```

```python
from sklearn.linear_model import LinearRegression
lin_reg = LinearRegression()
lin_reg.fit(X_train, y_train)
print('Współczynnik kierunkowy: ', lin_reg.coef_)
print('Współczynnik wyraz wolny: ', lin_reg.intercept_)
```

```python
plt.figure(figsize=(10,6))
plt.scatter(X_train, y_train, label='Dane treningowe')
plt.plot(X_train, lin_reg.predict(X_train), color='red', label='Prosta regresji liniowej')
plt.xlabel('X')
plt.ylabel('y')
plt.legend()
```

```python
plt.figure(figsize=(10,6))
plt.scatter(X_test, y_test, label='Dane testowe')
plt.plot(X_test, lin_reg.predict(X_test), color='red', label='Prosta regresji liniowej')
plt.xlabel('X')
plt.ylabel('y')
plt.legend()
```

```python
lin_reg.score(X_test, y_test)
```
