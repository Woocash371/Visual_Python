```python
import numpy as np
import pandas as pd
import plotly_express as px
import plotly.graph_objects as go
```

```python
y_true = 100 + 20 * np.random.randn(50)
y_true
```

```python
y_pred = y_true + 10 * np.random.randn(50)
y_pred
```

```python
result = pd.DataFrame({'y_true': y_true, 'y_pred': y_pred})
result.head()
```

```python
result['error'] = result['y_true'] - result['y_pred']
result.head()
```

```python
def plot_regression_results(y_true, y_pred):
    result = pd.DataFrame({'y_true': y_true, 'y_pred': y_pred})
    min = result[['y_true', 'y_pred']].min().min()
    max = result[['y_true', 'y_pred']].max().max()
    fig = go.Figure(data=[go.Scatter(x=result['y_true'], y=result['y_pred'], mode='markers'),
                          go.Scatter(x=[min, max], y=[min, max])],
                    layout=go.Layout(showlegend=False, width=500,height=500, xaxis_title='y_true', yaxis_title='y_pred', title='regression results'))
    fig.show()
plot_regression_results(y_true, y_pred)
```

```python
def mean_absolute_error(y_true, y_pred):
    return abs(y_true - y_pred).sum() / len(y_true)
mean_absolute_error(y_true, y_pred)
```

```python
from sklearn.metrics import mean_absolute_error
mean_absolute_error(y_true, y_pred)
```

```python
def mean_squared_error(y_true, y_pred):
    return ((y_true - y_pred)**2).sum() / len(y_true)
mean_squared_error(y_true, y_pred)
```

```python
from sklearn.metrics import mean_squared_error
mean_squared_error(y_true, y_pred)
```

```python
def root_mean_squared_error(y_true, y_pred):
    return np.sqrt(mean_squared_error(y_true, y_pred))
root_mean_squared_error(y_true, y_pred)
```

```python
np.sqrt(mean_squared_error(y_true, y_pred))
```

```python
def max_error(y_true, y_pred):
    return abs(y_true - y_pred).max()
```

```python
max_error(y_true, y_pred)
```

```python
from sklearn.metrics import max_error
max_error(y_true, y_pred)
```

```python
from sklearn.metrics import r2_score
r2_score(y_true, y_pred)
```

```python
def r2_score(y_true, y_pred):
    numerator = ((y_true - y_pred)**2).sum()
    denominator = ((y_true - y_true.mean())**2).sum()
    try:
        r2 = 1 - numerator / denominator
    except ZeroDivisionError:
        print('ZeroDivisionError')
    return r2
```

```python
r2_score(y_true, y_pred)
```
