```python
def rysowanie_3_wykresow(os_x = df_1_1['t'], os_y1 = df_1_1['m'], os_y2 = df_1_1['m'], 
                         os_y_p1 = df_1_1["Pads"], os_y_p2 = df_1_1["Pevap"],
                         os_y_p3 = df_1_1["Pevap"],os_y_p4 = df_1_1["Pads"],
                         os_y_zawory = df_1_1["V"],
                         
                         label = "Nazwa przebiegu 1", label2 = "Nazwa przebiegu 2",
                         label_p1 = "Nazwa Przebiegu Ciśnienia", label_p2 = "Nazwa Przebiegu Ciśnienia",
                        label_p3 = "Nazwa Przebiegu Ciśnienia", label_p4 = "Nazwa Przebiegu Ciśnienia",
                        zapis = "n"):
    
    
    plt.figure(figsize = (10,12))
    plt.subplot(3, 1, 1)
    plt.plot(os_x, os_y1, 
         label = label)
    plt.plot(os_x, os_y2, label = label2)
    plt.ylabel('masa pochłoniętej pary - m')
    plt.yticks(np.arange(0, 2, 0.2))
    plt.xticks(np.arange(0, 3100, 200))
    plt.grid()
    plt.legend()
    plt.legend(loc='upper center', bbox_to_anchor=(0.5, 1.05),
          ncol=3, fancybox=True, shadow=True)

    plt.subplot(3, 1, 2)
    plt.plot(os_x, os_y_p1,  label = label_p1 )
    plt.plot(os_x, os_y_p2,  label = label_p2 )
    plt.plot(os_x, os_y_p3,  label = label_p3 )
    plt.plot(os_x, os_y_p4,  label = label_p4 )
    plt.xticks(np.arange(0, 3100, 200))
    plt.yticks(np.arange(8, 36, 2))
    plt.ylabel('Ciśnienie - p [Pa]')
    plt.grid()
    plt.legend(loc='best' , ncol=3, fancybox=False, shadow=False)

    
    plt.subplot(3, 1, 3)  
    plt.plot(os_x, os_y_zawory)
    plt.xticks(np.arange(0, 3100, 200))
    plt.yticks(np.arange(0, 600, 500))
    plt.grid()
    plt.xlabel("Czas = t [s]")
    
    if zapis == "y":
        plt.savefig(input("Wpis nazwę pod jaką chcesz zapisać: "), dpi = 150)
        plt.show()
    else:
        plt.show()
```
