```python
def liczenie_miar(x):
    obliczone = []
    for seria in x:
        sr = seria.mean()
        pierwszy_element = x[0]
        sr_odniesienia = pierwszy_element.mean()
        roznica_sr = sr - sr_odniesienia

        mini = seria.min()
        mini_odniesienia = pierwszy_element.min()
        roznica_min = mini - mini_odniesienia

        maxi = seria.max()
        maxi_odniesnienia = pierwszy_element.max()
        roznica_max = maxi - maxi_odniesnienia

        mediana = seria.mean()
        mediania_odniesienia = pierwszy_element.mean()
        roznica_mediana = mediana - mediania_odniesienia


        obliczone.append([seria.name ,roznica_sr, roznica_min, roznica_max, roznica_mediana])

    df_obliczone = pd.DataFrame(obliczone)
    df_obliczone.columns = ('Parametr m','Różnica średniej', 'Różnica minimum', 'Różnica maximum', 'Różnica_mediana')
    return df_obliczone
```

```python
def liczenie_roznic_stat(x, wartosc_sr = 0.05):
    obliczone = []
    for seria in x:
        sr = seria.mean()
        pierwszy_element = x[0]
        sr_odniesienia = pierwszy_element.mean()
        roznica_sr = sr - sr_odniesienia
        if roznica_sr > wartosc_sr:
            mini = seria.min()
            mini_odniesienia = pierwszy_element.min()
            roznica_min = mini - mini_odniesienia

            maxi = seria.max()
            maxi_odniesnienia = pierwszy_element.max()
            roznica_max = maxi - maxi_odniesnienia

            mediana = seria.mean()
            mediania_odniesienia = pierwszy_element.mean()
            roznica_mediana = mediana - mediania_odniesienia


            obliczone.append([seria.name ,roznica_sr, roznica_min, roznica_max, roznica_mediana])

    df_obliczone = pd.DataFrame(obliczone)
    df_obliczone.columns = ('Parametr m','Różnica średniej', '\u0394 minimum', '\u0394 maximum', '\u0394 mediana')
    return df_obliczone
```
