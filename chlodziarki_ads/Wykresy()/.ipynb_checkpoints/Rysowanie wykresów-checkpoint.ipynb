```python
def wykres_masy(os_x=df_1_1["t"], os_y1=df_1_1["m"], os_y2=df_1_1["m"],
                zapis_wykresu="n", label2="wykres"):
    plt.figure(figsize=(10, 8))
    plt.plot(os_x, os_y1, label='m')
    plt.plot(os_x, os_y2, label=label2)
    plt.ylabel('masa pochłoniętej pary - m')
    plt.yticks(np.arange(0, 2, 0.2))
    plt.xticks(np.arange(0, 3100, 200))
    plt.xlabel(' Czas - t')
    plt.grid()
    plt.legend(loc='upper center', bbox_to_anchor=(0.5, 1.05),
               ncol=3, fancybox=True, shadow=True)
    if zapis_wykresu == "y":
        plt.savefig(input("Wpis nazwę pod jaką chcesz zapisać: "), dpi=150)
        plt.show()
    else:
        plt.show()
```

```python
def wykres_cisnienia(os_x = df_1_1["t"], os_y1 = df_1_1["Pads"] , os_y2 = df_1_1["Pads"], os_y3 = df_1_1["Pevap"], 
                     os_y4 = df_1_1["Pevap"], legenda1 = "NaN" , legenda2 = "NaN", legenda3 = "NaN", legenda4 = "NaN"
):
    plt.figure(figsize = (10,8))
    plt.plot(os_x, os_y1, label = legenda1)
    plt.plot(os_x, os_y2, label = legenda2)
    plt.plot(os_x, os_y3, label = legenda3)
    plt.plot(os_x, os_y4, label = legenda4)
    plt.xticks(np.arange(0, 3100, 200))
    plt.yticks(np.arange(8, 30, 2))
    plt.ylabel('Ciśnienie - p')
    plt.grid()
    plt.legend(loc='upper center', bbox_to_anchor=(0.5, 1.05),
          ncol=3, fancybox=True, shadow=True)
    
```
