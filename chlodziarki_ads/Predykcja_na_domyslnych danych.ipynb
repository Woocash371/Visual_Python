```python
import pandas as pd 
import numpy as np
import matplotlib.pyplot as plt

def drawing_plots(x):
    m_zero = x['masa mieszankki/15 [g/g]']
    m = x['m_PREDICTION']
    t = x['Czas [s]']
    Pads = x['P2 [mbar]']
    Pevap = x['P1 [mbar]']
    temp = x['T3 [C]']
    temp2 = x['T5 [C]']
    IDa = x['IDa'].unique()
    IDc = x['IDc'].unique()
    os_t = np.arange(0, 1600, 100)
    percentage = IDc

    if IDa == 1:
        material = 'SilicaGel '
    elif IDa == 2:
        material = 'SilicaGel+AL '
    elif IDa == 3:
        material = 'SilicaGel+NC '


    f, (ax1, ax2, ax3) = plt.subplots(3,1)
    f.set_figwidth(10)
    f.set_figheight(8)
    ax1.set_title('Wykres ' + material + percentage)
    ax1.plot(t, m_zero, label = 'wartosci_doswiadczalne ' + material + percentage)
    ax1.plot(t, m, label =  'predykcja '+ material + percentage)
    ax1.legend()
    ax1.grid(visible = True)
    ax1.set_xticks(os_t)
    
    ax2.plot(t, Pads)
    ax2.plot(t, Pevap)
    ax2.grid(visible=True)
    ax2.set_xticks(os_t)
        
    ax3.plot(t, temp, label = "T3 [C]")
    ax3.plot(t, temp2, label = 'T5 [C]')
    ax3.grid(visible=True)
    ax3.set_xticks(os_t)
    ax3.legend()
```

```python
df = pd.read_csv('Bez_zmian_walidacja.csv')
df
```

```python
list_of_experiment = df['numer badania'].unique()
dataFrame_list = []
```

```python
for element in list_of_experiment:
    dataFrame_list.append(df[df['numer badania'] == element])
```

```python
for element in dataFrame_list:
    drawing_plots(element)
```
